### GroupBy 客戶 與 消費類別

```python
chid_groupby_shop_tag = df.groupby(['chid', 'shop_tag'])
```

```python
tmp = chid_groupby_shop_tag[['txn_cnt', 'txn_amt']].sum()
cnt_amt_df = tmp.sort_values('txn_cnt', ascending= False).sort_index(level='chid', sort_remaining=False)
cnt_amt_df
```
